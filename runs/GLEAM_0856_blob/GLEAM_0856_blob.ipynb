# Running ZFINDER on GLEAM_0856

```python
from zfinder import zfinder

fitsfile = '0856_cube_c0.4_nat_80MHz_taper3.fits'
ra = '08:56:14.9' # small blobby field
dec = '02:24:05.1'
aperture_radius = 3
transition = 115.271

source = zfinder(fitsfile, ra, dec, aperture_radius, transition)
```

## Template Fitting Method

```python
z_start = 5
dz = 0.0001
z_end = 6

source.template(z_start, dz, z_end)
```

## FFT Fitting Method

```python
z_start = 5
dz = 0.0001
z_end = 6

source.fft(z_start, dz, z_end)
```

## Template Per Pixel Method

```python
size = 11
z_start = 5.50
dz = 0.0001
z_end = 5.55
aperture_radius_pp = 0.5

source.template_pp(size, z_start, dz, z_end, aperture_radius_pp)
```

## FFT Per Pixel Method

```python
size = 11
z_start = 5.50
dz = 0.0001
z_end = 5.55
aperture_radius_pp = 0.5

source.fft_pp(size, z_start, dz, z_end, aperture_radius_pp)
```

## Replot FFT Per Pixel Method

```python
from zfinder import Plotter

plot = Plotter()
plot.plot_heatmap_fromcsv(filename='fft_per_pixel.csv', subsize=7, flux_limit=0.001, contfile='0856.cont_map.fits')
```
